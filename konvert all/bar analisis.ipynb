```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
bro = pd.read_csv('ini bro_plotting.csv')
```

```python
hapus_bro = 'Unknown ship type'
filtered_bro = bro[bro['Ship_Type'] != hapus_bro]
```

```python
print(filtered_bro)
```

```python
bro_gabungkan = filtered_bro.groupby('Ship_Type').size()
```

```python
bro_gabungkan.plot(kind='bar', figsize=(10, 6), color='skyblue')
plt.xlabel('Ship Type')
plt.ylabel('Number of Ships')
plt.title('Number of Ships by Type')
plt.xticks(rotation=45)
plt.show()
```

```python
selat_madura_coordinates = [(-7.1056, 112.8428)]
```

```python
jenis_kapal_yang_diinginkan = 'Cargo ships'  
filtered_data = bro[bro['Ship_Type'] == jenis_kapal_yang_diinginkan]
```

```python
plt.figure(figsize=(10, 6))
plt.plot(filtered_data['Longitude'], filtered_data['Latitude'], marker='o', linestyle='-', label=jenis_kapal_yang_diinginkan)
for coord in selat_madura_coordinates:
    plt.plot(coord[1], coord[0], marker='x', color='red', markersize=10)
    plt.text(coord[1], coord[0], 'Selat Madura', fontsize=12, color='red')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(filtered_data['Longitude'], filtered_data['Latitude'], marker='o', linestyle='-', label=jenis_kapal_yang_diinginkan)

plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title(f'Rute Kapal untuk Jenis Kapal: {jenis_kapal_yang_diinginkan}')
plt.legend()
plt.show()
```
